```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función a integrar
def f(x):
    return np.exp(x)

# Regla compuesta de Simpson
def regla_simpson_compuesta(f, a, b, n):
    """
    Implementa la regla compuesta de Simpson para calcular la integral de f en [a, b]
    con n subintervalos (n debe ser par).
    """
    if n % 2 != 0:
        raise ValueError("n debe ser par")

    h = (b - a) / n  # Tamaño de paso
    x = np.linspace(a, b, n + 1)  # Puntos x0, x1, ..., xn
    y = f(x)

    # Aplicación de la fórmula de Simpson compuesta
    S = y[0] + y[-1]  # f(x0) + f(xn)
    S += 4 * np.sum(y[1:n:2])  # f(x1) + f(x3) + ...
    S += 2 * np.sum(y[2:n-1:2])  # f(x2) + f(x4) + ...

    return (h / 3) * S

# Parámetros del ejemplo
a = 0
b = 4
n = 4  # Número de subintervalos

# Calcular la integral usando la regla de Simpson compuesta
aprox = regla_simpson_compuesta(f, a, b, n)
real = np.exp(4) - np.exp(0)  # Solución exacta: e^4 - e^0
error = abs(real - aprox)

# Mostrar resultados
print(f"Aproximación con regla de Simpson compuesta: {aprox:.8f}")
print(f"Valor exacto: {real:.8f}")
print(f"Error absoluto: {error:.2e}")

# Graficar la función y los puntos usados en la regla
x_vals = np.linspace(a, b, 400)
y_vals = f(x_vals)

x_nodes = np.linspace(a, b, n + 1)
y_nodes = f(x_nodes)

plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='f(x) = e^x', color='red')
plt.plot(x_nodes, y_nodes, 'bo', label='Nodos de Simpson')
plt.fill_between(x_vals, y_vals, color='orange', alpha=0.2)
plt.title("Aplicación de la Regla de Simpson compuesta")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
