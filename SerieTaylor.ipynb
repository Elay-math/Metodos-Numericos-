$Gráfica$ $de$ $la$ $función$ $coseno$

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
  return np.cos(x)
x=np.linspace(-5, 5,400)
y=f(x)
plt.plot(x,y,label=r"$f(x) = \cos(x)$", color="b")
plt.xlabel=("x")
plt.ylabel=("y")
plt.title=("Cosine funtion")
plt.legend()
plt.grid(True)
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
def fac(a):
  if a==0:
    return 1
  else:
    return a*fac(a-1)



def f(x):
  return 1-(x**2)/fac(2)+(x**4)/fac(4)-(x**6)/fac(6)
x=np.linspace(-5, 5,1000)
y=f(x)
plt.plot(x,y,label=r"$f(x) = \lambda(x)$", color="b")
plt.xlabel=("x")
plt.ylabel=("y")
plt.title=("Taylor funtion")
plt.grid(True)
plt.legend()
plt.show()
```
