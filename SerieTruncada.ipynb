Serie Truncada de Taylor

Definimos la serie truncada de Taylor como:
$$
f(x) = \sum_{n=0}^{k+1} \frac{f^{(n)}(x_0)}{n!} (x - x_0)^n,
$$
para hacer aproximaciones en términos de polinomios a diferentes funciones. Se define un punto $x0$ donde se centrará la serie y $k$ el grado del polinomio que se desea obtener.

En este ejercicio se aproximará la función $ln(x)$ con la serie truncada de Taylor de grado 4

```python
#Importamos las librerias
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
#Definimos la variable x
x0=1
x=sp.symbols("x")
x
```

```python
#Definimos la función logaritmo natural que es con la cual trabajaremos
f=sp.log(x)
f
```

```python
#Definimos la serie de Taylor de grado 4 centrada en x0=1
Ts=sp.series(f,x,x0,5).removeO()
Ts
```

```python
#mandamos la función del logaritmo y la serie de taylor a funciones numericas
f_num=sp.lambdify(x,f,'numpy')
Ts_num=sp.lambdify(x,Ts,'numpy')
```

```python
#Creamos el intervalo y las particiones
xx=np.linspace(x0,x0+1,1000)
#Evaluamos las funciones en ese intervalo
yc=np.log(xx)
yf=Ts_num(xx)
```

```python
#Graficamos las funciones
plt.plot(xx,yc,label='ln(x)')
plt.plot(xx,yf,label='Serie de Taylor orden 4',color='red',linestyle='--')
plt.legend()
plt.title('Serie de Taylor de $f(x)=ln(x)$')
plt.grid(True)
```
